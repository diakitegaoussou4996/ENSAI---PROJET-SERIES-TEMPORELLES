```python
# Remplacer les virgules par des points et convertir en num√©rique
df_conso_brut['Consommation Brut (MW)'] = pd.to_numeric(df_conso_brut['Valeur (GW)'].str.replace(',', '.'), errors='coerce') * 1000
# Arrondir la colonne 'Consommation Brut (MW) ' et convertir en entiers
df_conso_brut['Consommation Brut (MW)'] = df_conso_brut['Consommation Brut (MW)'].round(0)

# Filtrer les donn√©es pour conserver uniquement les lignes avec "Consommation brute" et supprimer la colonne "Fili√®re"
df_conso_brut = df_conso_brut[df_conso_brut["Fili√®re"] == "Consommation brute"].drop(['Fili√®re', 'Valeur (GW)'], axis=1)
df_conso_brut = df_conso_brut.groupby('Date').mean().reset_index()
df_conso_brut.head()
```

<div style="font-family: 'Arial', sans-serif; color: #333; font-size: 14px;">
    <h3 style="color: #339966;">Analyse de la Stationnarit√© des S√©ries Temporelles</h3>
    <p>
        L'objectif de cette partie est de d√©terminer la stationnarit√© de la s√©rie temporelle de la consommation √©nerg√©tique brute. La stationnarit√© est un attribut crucial des s√©ries temporelles qui affecte la s√©lection des mod√®les pr√©dictifs et la validit√© des pr√©visions.
    </p>
    <p>
        Pour √©valuer la stationnarit√©, nous appliquons trois tests statistiques : 
    </p>
    <ul>
        <li>
            <strong>Le test Augmented Dickey-Fuller (ADF)</strong> examine l'hypoth√®se nulle selon laquelle une s√©rie poss√®de une racine unitaire, indiquant une possible non-stationnarit√©.
        </li>
        <li>
            <strong>Le test Phillips-Perron (PP)</strong> teste √©galement la pr√©sence d'une racine unitaire dans la s√©rie, mais il est robuste face √† de nombreuses formes de d√©pendance s√©rielle et de tendances d√©terministes.
        </li>
        <li>
            <strong>Le test Kwiatkowski-Phillips-Schmidt-Shin (KPSS)</strong> a pour hypoth√®se nulle la stationnarit√© autour d'une tendance, avec l'alternative √©tant la non-stationnarit√©.
        </li>
    </ul>
    <p>
        Les r√©sultats combin√©s de ces tests nous permettent d'√©tablir une conclusion sur la stationnarit√© de la s√©rie temporelle. Une proc√©dure de d√©cision bas√©e sur les valeurs p des tests nous aide √† d√©terminer si la s√©rie est <em>stationnaire</em> ou <em>non-stationnaire</em>. Ces informations sont essentielles pour le d√©veloppement de mod√®les pr√©visionnels fiables et pour la planification strat√©gique dans le secteur √©nerg√©tique.
    </p>
    <p>
        Les r√©sultats de l'analyse de stationnarit√© sont r√©capitul√©s dans un tableau pour une interpr√©tation claire. En cas de r√©sultats indiquant une non-stationnarit√©, des √©tapes de diff√©renciation ou de transformation des donn√©es peuvent √™tre n√©cessaires avant de proc√©der √† la mod√©lisation pr√©dictive.
    </p>
</div>


```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller, kpss
from arch.unitroot import PhillipsPerron

def stationnarite(data, var):
    sig = 0.05  # Seuil de signification
    res_df = pd.DataFrame()  # Cr√©er un DataFrame vide
    
    # Test ADF
    adf_test = adfuller(data[var], autolag='AIC')
    
    # Test Phillips-Perron
    pp_test = PhillipsPerron(data[var])
    
    # Test KPSS
    kpss_test = kpss(data[var], nlags="auto", regression='c')
    
    # Interpr√©tation des r√©sultats
    flag = 'Ind√©termin√©'
    if (adf_test[1] < sig) and (pp_test.pvalue < sig) and (kpss_test[1] > sig):
        flag = 'Stationnaire'
    elif (adf_test[1] > sig) or (pp_test.pvalue > sig) or (kpss_test[1] < sig):
        flag = 'Non-stationnaire'
    
    new_row = pd.Series({
        'Variable': var.upper(),
        'Statistique ADF': adf_test[0],
        'p-valeur ADF': adf_test[1],
        'Statistique PP': pp_test.stat,
        'p-valeur PP': pp_test.pvalue,
        'Statistique KPSS': kpss_test[0],
        'p-valeur KPSS': kpss_test[1],
        'Drapeau de stationnarit√©': flag
    })

    res_df = pd.concat([res_df, new_row.to_frame().T], ignore_index=True)
    
    return res_df

# Utilisation de l'exemple (√† ex√©cuter avec vos donn√©es r√©elles) :
# Remplacez 'your_dataframe' par votre DataFrame et 'Consommation Brut (MW)' par le nom de votre colonne
resultats_stationnarite = stationnarite(Conso_Energie, 'Consommation Brut (MW)')
resultats_stationnarite
```

Notre objectif est de trouver un mod√®le qui capture au mieux les caract√©ristiques intrins√®ques de nos donn√©es, y compris les tendances et la saisonnalit√©, afin de faire des pr√©visions pr√©cises pour l'avenir.
    </p>
    <p>
        Pour ce faire, nous utiliserons la fonction <code>auto_arima</code> de la biblioth√®que <code>pmdarima</code> en Python, qui va nous permettre de tester automatiquement plusieurs combinaisons de param√®tres pour les mod√®les ARIMA et SARIMA. Ce processus implique la recherche du meilleur ensemble de param√®tres (p, d, q) pour le mod√®le ARIMA et (P, D, Q, m) pour la composante saisonni√®re SARIMA, qui minimisent le crit√®re d'information d'Akaike (AIC) ou le crit√®re d'information bay√©sien (BIC), indiquant un meilleur ajustement du mod√®le.
    </p>
    <p>
        <ul>
            <li><code>start_p</code> et <code>start_q</code> indiquent les points de d√©part de la recherche pour les ordres du mod√®le AR et MA.</li>
            <li><code>max_p</code> et <code>max_q</code> d√©finissent les limites maximales pour les ordres du mod√®le AR et MA.</li>
            <li><code>m</code> repr√©sente la p√©riodicit√© saisonni√®re des donn√©es.</li>
            <li><code>start_P</code>, <code>d</code>, <code>D</code> et les options saisonni√®res permettent de tester les composantes saisonni√®res du mod√®le.</li>
            <li><code>trace</code> active l'affichage des √©tapes de la recherche, offrant une transparence dans la s√©lection du mod√®le.</li>
            <li><code>stepwise</code> active l'optimisation pas √† pas qui acc√©l√®re la recherche du meilleur mod√®le.</li>
        </ul>
    </p>
    <p>
        √Ä la fin du processus, nous obtiendrons un r√©sum√© d√©taill√© du meilleur mod√®le trouv√©. Ce r√©sum√© nous donnera des informations pr√©cieuses sur les param√®tres s√©lectionn√©s ainsi que sur les statistiques de performance du mod√®le, nous permettant de juger de l'ad√©quation du mod√®le choisi pour nos besoins de pr√©vision.
    </p>
</div>

```python
from pmdarima import auto_arima
import numpy as np
import pandas as pd
import random

# Fixer la graine al√©atoire pour la reproductibilit√©
np.random.seed(123)
random.seed(123)

# Assurez-vous que la colonne 'Consommation_Diff' est votre s√©rie temporelle
# et que 'Conso_Energie_Diff_Clean' est votre DataFrame
# Remplacer 'Conso_Energie_Diff_Clean' avec votre DataFrame r√©el
series = Conso_Energie_Diff_Clean['Consommation_Diff'].dropna()  # √âliminer les valeurs NaN pour √©viter les erreurs lors de la mod√©lisation

# Utiliser auto_arima pour trouver le meilleur mod√®le ARIMA
model = auto_arima(series,
                   start_p=0, start_q=0,  # Points de d√©part initiaux pour p et q
                   max_p=5, max_q=5,      # Limites sup√©rieures pour p et q
                   m=12,                  # Fr√©quence saisonni√®re des donn√©es
                   start_P=0,             # Point de d√©part initial pour P
                   seasonal=True,         # Indique si le mod√®le doit inclure des composants saisonniers
                   d=None, D=None,           # d est la diff√©renciation non saisonni√®re, D la saisonni√®re
                   trace=True,            # Affiche le d√©roulement de la recherche des meilleurs param√®tres
                   error_action='ignore', # Ignore les ordres qui ne convergent pas lors de la recherche
                   suppress_warnings=True,# Supprime les avertissements pour un affichage propre
                   stepwise=True,         # Utilise l'algorithme stepwise pour une recherche efficace
                   random_state=123)      # Permet la reproductibilit√© des r√©sultats de recherche

# Affiche le r√©sum√© du mod√®le pour examiner les param√®tres s√©lectionn√©s
#print(model.summary())

# Enregistrement des param√®tres du mod√®le pour l'initialisation future
best_params = model.get_params()

# Initialisation d'un mod√®le ARIMA avec les param√®tres obtenus
# Cela permet d'assurer la coh√©rence lors de la r√©utilisation du mod√®le
# Par exemple : ARIMA(p=best_params['order'][0], d=best_params['order'][1], q=best_params['order'][2])
```

<p style="font-size: 16px;">
            L'int√©gration de lags dans notre mod√®le permet d'encapsuler les effets de la consommation d'√©nergie des jours pr√©c√©dents sur les jours suivants. Ces lags agissent comme des √©chos temporels qui capturent non seulement les r√©ponses directes √† des stimuli comme les changements de temp√©rature et les √©v√©nements saisonniers, mais aussi les comportements diff√©r√©s tels que les ajustements d'usage apr√®s des jours f√©ri√©s ou des p√©riodes de pic.

Cette approche permet de d√©tecter les tendances globales et la variabilit√© dans la s√©rie temporelle, en r√©duisant l'impact des fluctuations journali√®res et en mettant en √©vidence les tendances sur une semaine.
</p>

<span style="font-size: 20px; margin-bottom: 5px; color: #4CAF50; font-weight: bold;">D√©composition de la S√©rie Temporelle</span>
<p style="font-size: 16px;">
    La d√©composition de la s√©rie temporelle en composantes de tendance, saisonni√®re et r√©siduelle offre une vue d√©taill√©e de la consommation. Cette m√©thode permet de s√©parer les effets √† long terme (tendance), les cycles r√©p√©titifs (saisonni√®re) et les variations inexpliqu√©es (r√©siduelle). Cela fournit des insights pr√©cieux pour la mod√©lisation et aide √† mieux comprendre les facteurs sous-jacents de la consommation.
</p>
    
   <span style="font-size: 20px; margin-bottom: 5px; color: #4CAF50; font-weight: bold;">Introduction des D√©lais dans les R√©sidus</span>
<p style="font-size: 16px;">
        En prenant en compte les r√©sidus, ou les parties de la consommation qui ne sont pas expliqu√©es par la tendance ou la saison, on peut d√©tecter des changements soudains ou des √©v√©nements impr√©vus, permettant au mod√®le de s'adapter rapidement.
 </p>
<p style="font-size: 16px;">
    Ensemble, ces caract√©ristiques avanc√©es enrichissent notre mod√®le pr√©dictif en offrant une compr√©hension plus nuanc√©e et d√©taill√©e de la consommation √©nerg√©tique. Elles permettent d'am√©liorer la pr√©cision des pr√©visions et de fournir des insights plus profonds sur la dynamique de la consommation d'√©nergie.
</p>
<p style="font-size: 16px;">
    Toutefois, il est essentiel de g√©rer la complexit√© accrue pour √©viter de surinterpr√©ter les r√©sultats ou d'introduire des erreurs telles que la multicollin√©arit√©.
</p>


```python
# Cr√©er des lag features
for lag in range(1, 5):
    Conso_Energie_Diff_Clean[f'Consommation_Diff_lag{lag}'] = Conso_Energie_Diff_Clean['Consommation_Diff'].shift(lag)
    
    # Cr√©er des rolling window features
window = 7  # Fen√™tre d'une semaine
Conso_Energie_Diff_Clean['Rolling_Mean'] = Conso_Energie_Diff_Clean['Consommation_Diff'].rolling(window).mean()
Conso_Energie_Diff_Clean['Rolling_Std'] = Conso_Energie_Diff_Clean['Consommation_Diff'].rolling(window).std()

# D√©composition de la s√©rie temporelle
decomposition = seasonal_decompose(Conso_Energie_Diff_Clean['Consommation_Diff'], model='additive')
Conso_Energie_Diff_Clean['Trend'] = decomposition.trend
Conso_Energie_Diff_Clean['Seasonal'] = decomposition.seasonal
Conso_Energie_Diff_Clean['Residual'] = decomposition.resid


# Cr√©er des lag Residual
for lag in range(1, 5):
    Conso_Energie_Diff_Clean[f'Residual{lag}'] = Conso_Energie_Diff_Clean['Residual'].shift(lag)
    

# Supprimer les NaN cr√©√©s par les lags et la fen√™tre glissante
Conso_Energie_Diff_Clean = Conso_Energie_Diff_Clean.dropna()

```

```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Assuming X_train and X_test are already loaded and are Pandas DataFrames

# Enregistrez l'index de y_test avant toute manipulation
test_dates = y_test.index

# Normalizing input features
scaler_x = MinMaxScaler(feature_range=(0, 1))
X_train_scaled = scaler_x.fit_transform(X_train)
X_test_scaled = scaler_x.transform(X_test)

# Normalizing output feature
scaler_y = MinMaxScaler(feature_range=(0, 1))
y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1))
y_test_scaled = scaler_y.transform(y_test.values.reshape(-1, 1))

# Reshaping input data for LSTM [samples, time steps, features]
X_train_reshaped = X_train_scaled.reshape((X_train_scaled.shape[0], 1, X_train_scaled.shape[1]))
X_test_reshaped = X_test_scaled.reshape((X_test_scaled.shape[0], 1, X_test_scaled.shape[1]))

# Define the LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train_reshaped.shape[1], X_train_reshaped.shape[2])))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dense(units=25))
model.add(Dense(units=1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Fit the model to the training data
model.fit(
    X_train_reshaped,
    y_train_scaled,
    epochs=100,
    batch_size=32,
    validation_data=(X_test_reshaped, y_test_scaled),
    verbose=1
)

# Predicting on the test set
y_test_pred_scaled = model.predict(X_test_reshaped)

# Inverse transforming the scaled prediction to actual values
y_test_pred = scaler_y.inverse_transform(y_test_pred_scaled)

# Create a DataFrame for the predictions using the original test dates index
y_test_pred_df = pd.DataFrame(y_test_pred, index=test_dates, columns=['Predictions'])

# Calculating RMSE for the test set
# Calculating RMSE and other metrics for both training and test set
test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred_df['Predictions']))

print(f"Test RMSE: {test_rmse}")
# Plotting the predictions alongside the actual values
plt.figure(figsize=(15, 7))
plt.plot(y_test, label='Actual values', color='blue')
plt.plot(y_test_pred_df['Predictions'], label='LSTM predicted values', color='red', alpha=0.7)
plt.title("Energy Consumption Prediction using LSTM")
plt.xlabel("Date")
plt.ylabel("Energy Consumption")
plt.legend()
plt.show()
```
